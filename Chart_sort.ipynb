```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("result.csv")

#plt.figure(figsize=(<width>, <height>)) -> thay đổi kích thước plot
```

# Thời gian

## Các thuật toán sắp xếp nhanh (Quick sort, Merge sort, Heap sort, Shell sort, Counting sort, Radix sort, Flash sort)

```python
filtered_data_quick = data[
    (data['Algorithm'] == 'quick-sort') & (data['Data Order'] == 'Randomize')
]
filtered_data_merge = data[
    (data['Algorithm'] == 'merge-sort') & (data['Data Order'] == 'Randomize')
]
filtered_data_heap = data[
    (data['Algorithm'] == 'heap-sort') & (data['Data Order'] == 'Randomize')
]
filtered_data_shell = data[
    (data['Algorithm'] == 'shell-sort') & (data['Data Order'] == 'Randomize')
]
filtered_data_counting = data[
    (data['Algorithm'] == 'counting-sort') & (data['Data Order'] == 'Randomize')
]
filtered_data_radix = data[
    (data['Algorithm'] == 'radix-sort') & (data['Data Order'] == 'Randomize')
]
filtered_data_flash = data[
    (data['Algorithm'] == 'flash-sort') & (data['Data Order'] == 'Randomize')
]

# input data
data_size = [10000, 30000, 50000, 100000, 300000, 500000]
quick_sort = filtered_data_quick['Running Time (ms)']
merge_sort = filtered_data_merge['Running Time (ms)']
heap_sort = filtered_data_heap['Running Time (ms)']
shell_sort = filtered_data_shell['Running Time (ms)']
counting_sort = filtered_data_counting['Running Time (ms)']
radix_sort = filtered_data_radix['Running Time (ms)']
flash_sort = filtered_data_flash['Running Time (ms)']

plt.plot(data_size, quick_sort, marker='o', linestyle='-', color='b', label='quick-sort')
plt.plot(data_size, heap_sort, marker='o', linestyle='-', color='r', label='heap-sort')
plt.plot(data_size, merge_sort, marker='o', linestyle='-', color='g', label='merge-sort')
plt.plot(data_size, shell_sort, marker='o', linestyle='-', color='y', label='shell-sort')

plt.plot(data_size, counting_sort, marker='o', linestyle='-', color='black', label='counting-sort')
plt.plot(data_size, radix_sort, marker='o', linestyle='-', color='aqua', label='radix-sort')
plt.plot(data_size, flash_sort, marker='o', linestyle='-', color='pink', label='flash-sort')


# Thêm tiêu đề và nhãn
plt.title("Biểu đồ thể hiện thời gian thực thi trên tập dữ liệu ngẫu nhiên (Randomize)")
plt.xlabel("Input")
plt.ylabel("Running time (ms)")
plt.legend()  # Hiển thị chú thích

# Hiển thị biểu đồ

plt.savefig('./img/time/fast/random.png', dpi=300)
plt.show()


```
